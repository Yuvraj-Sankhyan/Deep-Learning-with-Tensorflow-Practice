# Pipeline exercise 

```python
import tensorflow as tf
```

```python
tf_dataset = tf.data.Dataset.list_files("../extraction_dump/reviews/*/*",shuffle=False)
tf_dataset
```

```python
for items in tf_dataset:
    print(items.numpy())
```

Extract review text from these files. Extract label from folder name

```python
import os
def Extraction(file_path):
    review = tf.io.read_file(file_path)
    label = tf.strings.split(file_path,os.path.sep)[-2]
    return (review,label)
```

```python
review_ds = tf_dataset.map(Extraction)
for review, label in review_ds:
    print(review.numpy()[:20], "*****Label*****",label.numpy())
```

```python
review_ds = review_ds.filter(lambda review,label: review != "")
for review, label in review_ds:
    print(review.numpy()[:20], "*****Label*****",label.numpy())
```

```python
review_one_line = tf.data.Dataset.list_files("../extraction_dump/reviews/*/*",shuffle=False).map(Extraction).filter(lambda review,label: review != "").shuffle(3)

for review, label in review_one_line:
    print(review.numpy()[:20], "*****Label*****",label.numpy())
```
